```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
#tabela normal padronizada
tabela_normal_padronizada = pd.DataFrame([],index=[round(i/100,2) for i in range(0,400,10)],
                                         columns = [round(i/100,2) for i in range(10)])

for index in tabela_normal_padronizada.index:
    for column in tabela_normal_padronizada.columns:
        z = np.round(float(index) + float(column),2)
        tabela_normal_padronizada.loc[index,column] = round(norm.cdf(z),4)
tabela_normal_padronizada.rename_axis('Z', axis=1, inplace =True)
tabela_normal_padronizada
```

```python
#exemplo motorista
media = 300.00
desvpad = 50.00
z_superior1 = (350-media)/desvpad
z_inferior1 = (250-media)/desvpad
z_superior2 = (500-media)/desvpad
z_inferior2 = (400-media)/desvpad
probabilidade1 = norm.cdf(z_superior1) - (1-norm.cdf(z_superior1))
probabilidade2 = norm.cdf(z_superior2) - norm.cdf(z_inferior2)
print(f'{round(probabilidade1,2)}, {round(probabilidade2,2)}')
```

```python
#exemplo vida útil
## durar entre 650 e 750 dias
media = 720 
desvpad = 30
z_superior=(750-media)/desvpad
z_inferior=(650-media)/desvpad
probabilidade1 = norm.cdf(z_superior) - norm.cdf(z_inferior)
## > 800 dias
z = (800-media)/desvpad
probabilidade2 = norm.cdf(-z)
## < 700 dias
z = (700-media)/desvpad
probabilidade3 = norm.cdf(z)
print(f'{round(probabilidade1,4)}, {round(probabilidade2,4)}, {round(probabilidade3,4)}')
```
