## Estimadores de minimo quadrados ordinarios


```python
n = df.shape[0]
n
```

```python
soma_y = df.IMC.sum()
soma_y
```

```python
soma_x = df.Peso.sum()
soma_x
```

```python
soma_x2 = df.Peso.apply(lambda x: x**2).sum()
soma_x2
```

```python
soma_y2 = df.IMC.apply(lambda x: x**2).sum()
soma_y2
```

```python
soma_xy = df.apply(lambda data: data.Peso * data.IMC, axis=1).sum()
soma_xy
```

```python
df.head()
```

```python
#calculo beta2
numerador = n * soma_xy - soma_x*soma_y
denominador = n* soma_x2 - (soma_x)**2
beta2 = numerador/denominador
beta2
```

```python
beta1 = df.IMC.mean() - beta2 * df.Peso.mean()
beta1
```

## Teste de significancia dos parâmetros

```python
s = np.sqrt(eqm)
s
```

```python
soma_desvios = df.Peso.apply(lambda x: (x- df.Peso.mean())**2).sum()
soma_desvios
```

```python
s_beta2 = s/ (np.sqrt(soma_desvios))
s_beta2
```

```python
from scipy.stats import t as t_student
```

```python
t_alfa = t_student.ppf(0.975,graus_de_liberdade)
t_alfa
```

```python
t = (beta2 - 0)/s_beta2
t
```

```python
resultado_regressao_banco.tvalues
```

```python
#teste de hipótese valor critico - estatistica teste
if (t < - t_alfa or t > t_alfa):
    print(f'Hipótese nula rejeitada, t : {t.round(4)}')
else:
        print(f'Hipótese nula aceita, t : {t.round(4)}')
```

```python
#teste hipotese pvalor
p_valor = 2*(t_student.sf(t,graus_de_liberdade))
p_valor
```

```python
p_valor < 0.05
```
